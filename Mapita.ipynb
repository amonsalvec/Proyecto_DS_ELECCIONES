```python
df['LATITUD'] = df['COMARCA'].map(lambda x: comarca_coords.get(x, (None, None))[0])
df['LONGITUD'] = df['COMARCA'].map(lambda x: comarca_coords.get(x, (None, None))[1])
```

```python
df
eliminar = ['COD_MUNICIPIO', 'MUNICIPIO', 'DISTRITO', 'SECCION', 'MESA', 'CANDIDATURAS', 'CENSO']
df_bueno= df.drop(columns=eliminar)
```

```python
df_2009 = df_bueno[df_bueno['ANYO'] == 2009].copy()
df_2014 = df_bueno[df_bueno['ANYO'] == 2014].copy()
df_2019 = df_bueno[df_bueno['ANYO'] == 2019].copy()
```

```python
df_agg = df_2009.groupby('COMARCA').agg({
    'VOTANTES': 'sum',
    'VALIDOS': 'sum',
    'BLANCOS': 'sum',
    'ABSTENCIONES': 'sum',
    'NULOS': 'sum',
    'LATITUD': 'mean',
    'LONGITUD': 'mean'
}).reset_index()

m = folium.Map(location=[39.5, -0.5], zoom_start=8)

for _, row in df_agg.iterrows():
    popup_text = (
        f"<b>{row['COMARCA']}</b><br>"
        f"Votantes: {row['VOTANTES']}<br>"
        f"Votos válidos: {row['VALIDOS']}<br>"
        f"Votos blancos: {row['BLANCOS']}<br>"
        f"Abstenciones: {row['ABSTENCIONES']}<br>"
        f"Votos nulos: {row['NULOS']}"
    )
    folium.Marker(
        location=[row['LATITUD'], row['LONGITUD']],
        popup=popup_text
    ).add_to(m)

m.save('mapa_comarcas_2009.html')
m
```
