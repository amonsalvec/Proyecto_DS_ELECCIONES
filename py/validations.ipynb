# Pruebas de aleatoriedad

Cuando los datos no provienen de una muestra aleatoria, el riesgo principal es que los resultados sean inherentemente sesgados. Esto afecta la capacidad de generalizar conclusiones a la población en estudio, pudiendo llevar a decisiones equivocadas basadas en una visión distorsionada de la realidad. Para detectar este problema y validarlo estadísticamente, puedes proceder de la siguiente manera:


1. Verificar la Aleatoriedad del Muestra
A. Pruebas Estadísticas de Aleatoriedad
- Prueba de Runs: Esta prueba detecta patrones en una secuencia de datos, evaluando si la secuencia presenta “corridas” (runs) que se desvíen de lo esperado en una secuencia puramente aleatoria. Si el número de runs resulta significativamente distinto a lo esperado, se puede sospechar de aleatoriedad defectuosa.
- Autocorrelación: Calcular coeficientes de autocorrelación para detectar si los datos presentan dependencias temporales o secuenciales. Una elevada autocorrelación puede indicar que la muestra ha sido recolectada a partir de un proceso sistemático o con algún orden implícito, en lugar de ser puramente aleatorio.
- Prueba Kolmogórov-Smirnov o Chi-cuadrado: Comparar la distribución empírica de los datos con la distribución teórica esperada para determinados parámetros de la población. Cualquier discrepancia significativa podría sugerir una falta de aleatoriedad.
B. Análisis de Metadatos y Procedimientos de Muestreo
- Revisión del Proceso de Recolección: Investigar si el protocolo de muestreo se realizó adecuadamente y sin sesgos preestablecidos. Por ejemplo, determinar si hubo un auto-seleccionamiento de participantes o si se usaron métodos de conveniencia en lugar de un muestreo probabilístico.
- Análisis de Subgrupos: Comparar estadísticas descriptivas entre diferentes subgrupos de la colección de datos. Una diferencia notoria podría evidenciar que ciertos segmentos de la población están sobre o subrepresentados.

2. Análisis del Contenido de los Conjuntos de Datos
A. Estadística Descriptiva y Visualización
- Medidas Centrales y Dispersión: Calcula la media, mediana, desviación estándar, rangos y percentiles para detectar asimetrías o diferencias inusuales en la distribución de los datos.
- Visualizaciones:
- Histogramas y Diagramas de Caja (Boxplots): Te permitirán identificar la forma de la distribución, posibles asimetrías y outliers.
- Gráficos de Dispersión y Q-Q Plots: Son útiles para evaluar la dispersión de datos y compararla con una distribución teórica (por ejemplo, la normalidad).
B. Exploración de Relaciones entre Variables
- Análisis de Correlación y Regresión: Establecer asociaciones entre variables. Si los patrones difieren significativamente de lo que sería esperado en una muestra aleatoria, puede ser un indicador de sesgo en los datos.
- Detección de Outliers y Valores Atípicos: Identificar observaciones que se desvíen enormemente del resto, lo cual podría ser consecuencia de errores en la recolección o de un sesgo en la selección de la muestra.

3. Validaciones de Carácter Estadístico
A. Validaciones Internas
- Bootstrap y Técnicas de Re-Muestreo: Estas técnicas permiten evaluar la estabilidad de las estimaciones obtenidas. Si al re-muestrear los datos se obtienen resultados muy variables, podría indicar falta de representatividad y robustez en la muestra.
- Cross-Validation (para modelos predictivos): Ayuda a confirmar si los patrones encontrados en una parte de la muestra se mantienen en otra, lo que es esencial cuando la muestra original podría tener sesgos.
B. Validaciones Externas
- Comparación con Otras Fuentes de Datos: Confronta tus hallazgos con estudios o bases de datos conocidos y representativos. Una gran discrepancia sustancial puede sugerir que tu muestra es defectuosa o está sesgada.
- Test de Hipótesis y Evaluación de Supuestos:
- Realiza tests de normalidad (como Shapiro-Wilk) para asegurar que los supuestos del modelo estadístico se cumplen.
- Verifica la homogeneidad de varianzas (por ejemplo, mediante el test de Levene) para asegurarte de que ninguna categoría evidencie una variabilidad extraordinaria que afecte las conclusiones.
C. Ajustes y Correcciones
- Reponderación (Weighting): Si identificas que ciertos subgrupos están sobre o subrepresentados, puedes aplicar técnicas de ponderación para ajustar la influencia de dichas observaciones.
- Análisis de Sensibilidad: Explora cómo varían los resultados al modificar ciertos parámetros del modelo o al excluir ciertos segmentos de datos, lo que ayuda a comprender el impacto que tiene el sesgo muestral.

Reflexión Final
Si los datos no proceden de una muestra aleatoria, cualquier inferencia o conclusión derivada estará sesgada, afectando la validez del estudio o el modelo predictivo. Por ello, es vital combinar técnicas tanto gráficas como estadísticas para detectar y, en la medida de lo posible, corregir estos sesgos. Esto no solo fortalece la credibilidad de los análisis, sino que además orienta sobre cómo plantear futuras recolecciones de datos de manera que sean representativas de la población en estudio.
Para profundizar en el tema, podrías explorar casos prácticos del uso de estas técnicas en software estadístico como R o Python, donde librerías como scipy, statsmodels y pandas ofrecen herramientas específicas para estas evaluaciones. Además, investigar métodos robustos para la detección de sesgos en conjuntos de datos en campos específicos, como la epidemiología o el análisis de mercados, te permitirá ajustar las metodologías al contexto en particular. 

```python
import numpy as np
from scipy import stats

# Datos generados aleatoriamente (supongamos que deberían seguir una distribución normal)
data = np.random.randn(100)

# Realizamos el test KS comparando con la distribución normal ('norm')
ks_stat, p_value = stats.kstest(data, 'norm')
print("Estadístico KS:", ks_stat)
print("p-valor:", p_value)

if p_value < 0.05:
    print("Los datos no se distribuyen como una normal")
else:
    print("Los datos se distribuyen como una normal")

# Resultados:
# Estadístico KS: 0.0
# p-valor: 0.0

# Los datos se distribuyen como una normal
```

```python
# Análisis de autocorrelación
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generamos una serie aleatoria
data = pd.Series(np.random.randn(100))

# Graficar la autocorrelación usando la función integrada de Pandas
pd.plotting.autocorrelation_plot(data)
plt.title("Autocorrelación de los datos")
plt.show()

# Calcular la autocorrelación para un lag específico, por ejemplo, lag=1:
lag1_ac = data.autocorr(lag=1)
print("Autocorrelación con lag 1:", lag1_ac)
```

Interpretación:
Si la autocorrelación para lags pequeños es alta, puede indicar que los 
datos presentan dependencias que se alejan de lo esperado en una muestra aleatoria.

```python
import numpy as np

def runs_test(data):
    """
    Realiza la prueba de corridas sobre una secuencia numérica.
    Se utiliza la mediana para convertir la serie en valores binarios (mayor o menor que la mediana).
    """
    median = np.median(data)
    runs = 1  # Se cuenta la primera corrida
    n1 = 0  # Número de valores por encima de la mediana
    n2 = 0  # Número de valores por debajo de la mediana

    # Determinar el primer valor
    previous = data[0] > median
    if previous:
        n1 += 1
    else:
        n2 += 1

    # Iterar sobre el resto de la secuencia
    for x in data[1:]:
        current = x > median
        if current != previous:
            runs += 1
            previous = current
        if current:
            n1 += 1
        else:
            n2 += 1

    # Cálculo de la cantidad esperada de corridas y su desviación estándar
    runs_expected = (2 * n1 * n2) / (n1 + n2) + 1
    runs_std = np.sqrt((2 * n1 * n2 * (2 * n1 * n2 - n1 - n2)) / (((n1 + n2) ** 2) * (n1 + n2 - 1)))
    z = (runs - runs_expected) / runs_std
    return runs, runs_expected, runs_std, z

# Ejemplo de uso:
data = np.random.randn(100)  # Datos generados aleatoriamente
runs, runs_expected, runs_std, z = runs_test(data)
print("Número de corridas:", runs)
print("Número esperado:", runs_expected)
print("Desviación estándar:", runs_std)
print("Estadístico Z:", z)

# Resultados 
# Interpretación:
#Un valor de z mayor en valor absoluto (por ejemplo, mayor a 1.96 
# para un nivel de significancia del 5%) sugiere que la secuencia 
# difiere significativamente de lo que se esperaría en un proceso 
# aleatorio.
```

Interpretación:
Un p-valor bajo (típicamente < 0.05) indica que es poco probable que los datos 
provengan de la distribución normal, lo que puede ser un indicativo de que la muestra 
tenga sesgos o que los supuestos no se cumplan.


```python
# Visualización de datos: Histogramas y Diagramas de Caja
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.figure(figsize=(12, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Histograma de los datos")

# Diagrama de caja (Boxplot)
plt.subplot(1, 2, 2)
plt.boxplot(data, vert=False)
plt.title("Diagrama de caja")
plt.show()
```

Interpretación:

```python
# Validación Interna con Técnicas de Re-muestreo (Bootstrap)
#El método bootstrap permite evaluar la estabilidad de medidas estadísticas (por ejemplo, la media) al re-muestrear los datos con reemplazo.
import numpy as np
import matplotlib.pyplot as plt

def bootstrap_mean(data, n_iterations=1000):
    means = []
    n = len(data)
    for _ in range(n_iterations):
        sample = np.random.choice(data, size=n, replace=True)
        means.append(np.mean(sample))
    return means

data = np.random.randn(100)
boot_means = bootstrap_mean(data, n_iterations=10000)
print("Media estimada por bootstrap:", np.mean(boot_means))

# Visualizamos la distribución de las medias obtenidas por bootstrap
plt.hist(boot_means, bins=30, color="salmon", edgecolor='black')
plt.title("Distribución Bootstrap de la Media")
plt.xlabel("Valor de la media")
plt.ylabel("Frecuencia")
plt.show()
```

Interpretación:
Una distribución muy amplia de la media bootstrap sugiere que la estimación es sensible a las variaciones en la muestra, lo que puede estar relacionado con un muestreo no representativo o con alta variabilidad en los datos.


```python
# Validación Externa: Cross-Validation con scikit-learn
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import numpy as np

# Simulamos datos para una regresión simple
np.random.seed(0)
X = np.random.randn(100, 1)  # Variable independiente
y = 2 * X.squeeze() + np.random.randn(100) * 0.5  # Variable dependiente con algo de ruido

# Modelo de regresión lineal
model = LinearRegression()
scores = cross_val_score(model, X, y, cv=5)  # Validación cruzada de 5 particiones
print("Scores de validación cruzada:", scores)
print("Score promedio:", np.mean(scores))
```

Interpretación:
Una variabilidad alta entre las particiones o un score promedio bajo puede ser indicativo de inestabilidad en los datos, lo cual puede derivar de un muestreo defectuoso o no representativo.
